```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
data1 = scaler.fit_transform(np.array(data1).reshape(-1,1))
```

#### Splitting the dataset into train and test split

For time series datasets, the current data is always dependent on your previous data. Today's stock price is dependent on yesterday's price. For this reason, we cannot devide the data into train & test sets in some random fashion. 

If you have a,b,c,d,e in your timeseries dataset, then we sould only divide them as train -> a,b,c and test -> d,e (Do not change the order)


```python
# I am setting the size of my training set to be 70% of my entire dataset
training_size = int(len(data1)*0.7)

# Test set will automatically be the rest
test_size = len(data1) - training_size

# Splitting the data
train_data, test_data = data1[0:training_size, :], data1[training_size:len(data1), :1]
```

```python
training_size, test_size
```

```python
import numpy

# Convert an array of values into a dataset matrix

#Time Step defines the number of data to which my current data should be depended on
def create_dataset(dataset, time_step = 1): 
    X_data, Y_data = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]  
        X_data.append(a)
        Y_data.append(dataset[i + time_step, 0])
    return numpy.array(X_data), numpy.array(Y_data)
```

```python
time_step = 100
X_train, Y_train = create_dataset(train_data, time_step)
X_test, Y_test = create_dataset(test_data, time_step)
```

```python
print(X_train.shape)
print(X_test.shape) 
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Reshaping input to be [samples, time steps, features] which is required for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] , 1)
```

```python
print(X_train.shape)
print(X_test.shape) 
```

### Creating a stacked LSTM model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()

#input_shape has to be (X_train.shape[1] , 1) from the reshaping code
model.add(LSTM(50, return_sequences = True, input_shape = (100,1)))  
model.add(LSTM(50, return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 100, batch_size = 64, verbose = 1)
```

```python
import tensorflow as tf
```

```python
# Lets Do the prediction and check performance metrics

train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
# Transforming back to original form

train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
```

```python
# Calculating the RMSE performance metrics

import math
from sklearn.metrics import mean_squared_error

print('Mean Squared Error (Train):{}'.format(math.sqrt(mean_squared_error(Y_train, train_predict))))
print('Mean Squared Error (Test):{}'.format(math.sqrt(mean_squared_error(Y_test, test_predict))))
```

### Visualization

```python
# Shifting the train predictions for plotting

look_back = 100
trainPredictPlot = numpy.empty_like(data1)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict) + look_back, :] = train_predict

# Shifting the test predictions for plotting

testPredictPlot = numpy.empty_like(data1)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(train_predict) + (look_back*2) + 1:len(data1)-1, :] = test_predict

# Plotting the baseline and predictions

plt.plot(scaler.inverse_transform(data1))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

```python
# Forecasting data for the next 10 days

from numpy import array

lst_output = []
n_steps = 100
i = 0

# First the loop will go to the else case as the temp_input has only 100 data.
# In the else case, we predict the 101th data and add it to Y_pred. Y_pred is also added to temp_input. 
# Now, temp_input has 101 data and so the if loop will be activated.
# In the if case, we move the data_inut from 2nd - 101st data and add it to X_input.
# After reshaping the data, we predict the 102nd data and is added to Y_pred and the Y_pred is added to temp_input.

while(i < 30):
    
    if(len(temp_input) > 100):
        x_input = np.array(temp_input[1:])
        print("{} day input {}".format(i, x_input))
        
        x_input = x_input.reshape(1, -1)
        x_input = x_input.reshape((1, n_steps, 1))     #Reshaping the input for LSTM
        Y_pred = model.predict(x_input, verbose = 0)
        print("{} day output {}".format(i, Y_pred))
        
        temp_input.extend(Y_pred[0].tolist())
        temp_input = temp_input[1:]
        lst_output.extend(Y_pred.tolist())
        i = i+1
        
    else:
        x_input = x_input.reshape((1, n_steps, 1))    #Reshaping the input for LSTM
        Y_pred = model.predict(x_input, verbose = 0)
        print(Y_pred[0])
        
        temp_input.extend(Y_pred[0].tolist())
        print(len(temp_input))
        lst_output.extend(Y_pred.tolist())
        i = i+1
    

print(lst_output)
```
